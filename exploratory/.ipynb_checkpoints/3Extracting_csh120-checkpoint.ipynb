```python
import numpy as np
import matplotlib.pyplot as plt
np.seterr(divide='warn', invalid='warn')
```

```python
from yellowbrick.features import ParallelCoordinates
plt.figure(figsize=(30,20))
visualizer = ParallelCoordinates(
    classes=classes, sample=0.1, shuffle=True,
    features=X_vis_time.columns, normalize='standard',
    alpha=.1, fast=True
)

# Fit and transform the data to the visualizer
visualizer.fit_transform(X_vis_time, y_vis)

# Finalize the title and axes then display the visualization
visualizer.show()
```

I would like to apply the above figure to a class balanced set in the future, this would make this visualization alot more useful. I'm going to make my train and test sets now so that I can oversample on the train

```python
from sklearn.model_selection import train_test_split
X, y = csh120.drop('activity', axis=1), csh120.activity
X_train, X_test, y_train, y_test = train_test_split(X, y)
```

```python
from imblearn.over_sampling import RandomOverSampler
ros = RandomOverSampler()
X_res, y_res = ros.fit_resample(X_train, y_train)
csh120_train_res = pd.concat([X_res, y_res], axis=1)
```

```python
from yellowbrick.features import ParallelCoordinates

classes = list(y.value_counts().index)[:6]
X_vis = X_res[y_res.isin(classes)]
le = LabelEncoder()
y_vis = y_res[y_res.isin(classes)]
le.fit(y_vis)
y_vis = le.transform(y_vis)
X_vis_time = X_vis.iloc[:, 25:35]
X_vis_counts = X_vis.iloc[:, 14:24]
X_vis_sensor_info = X_vis.iloc[:, :12]
```
